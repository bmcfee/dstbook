# Exercises

```{exercise}
:label: ex4-1

How many complex numbers $z$ are there, such that:

1. $z^2 = 1$?
1. $z^2 = -1$?
1. $z^3 = 1$?
```

```{exercise}
:label: ex4-2

For each pair of numbers below, compute the sum $z+w$ and product $z\cdot w$.  Check your results with Python code.

1. $z=\mathsf{j}, w= 1/2 + \mathsf{j}$
1. $z=2 + 3\mathsf{j}, w= 3 - 2\mathsf{j}$
1. $z=e^{\mathsf{j}\cdot \pi}, w = e^{-3\mathsf{j}\cdot \pi/2}$
```

```{exercise}
:label: ex4-3

What are the complex conjugates of the following numbers:

1. $z = 1 + 2\mathsf{j}$?
1. $z=3$?
1. $z = 3e^{\mathsf{j}\cdot \pi}$?
1. $z = 3e^{\mathsf{j}\cdot \pi/2}$?
```

```{exercise}
:label: ex4-4

Let $z = e^{\mathsf{j}\cdot \pi/3}$, and consider the sequence generated by raising it to successive powers: $z^0, z^1, z^2, \dots, z^n, \dots$

1. What is the smallest power $n>1$ that satisfies $z^n = z$?
1. If we take $f_s = 100$, can you find a frequency $f > 0$ such that $\cos\left(2\pi \cdot f \cdot n / f_s\right) = \mathsf{Re}\left[z^n\right]$ (the real part of $z^n$) for all $n$?
```
